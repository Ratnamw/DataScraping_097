# SELENIUM

```python
import pandas as pd
from bs4 import BeautifulSoup

import chromedriver_autoinstaller
from selenium import webdriver
```

```python
#INSTALL BROWSER SUCH AS CHROME OR EDGE
chromedriver_autoinstaller.install()

#USE THE INSTALLED BROWSER FOR SELENIUM
browser = webdriver.Chrome()
```

```python
#IT IS SIMILAR TO REQUESTS.GET
# (SIMPLY TO GET THE WEBSITE THAT WE'RE TRYIN TO SCRAPE)
browser.get("https://www.thetimes.co.uk/sunday-times-rich-list")
```

```python
#MAKE IT BEAUTIFUL HTML
soup = BeautifulSoup(browser.page_source,"lxml")
```

```python
soup
```

```python
element_td = soup.find_all('td')
```

```python
element_td
```

```python
data_quote = []

for td_quote in element_td:
    data_quote.append(td_quote)
```

```python
data_quote
```

```python
data_rank =[]

for r in range(30, len(data_quote),5):
    data_rank.append(data_quote[r].text)
```

```python
data_rank
```

```python
rank_fix = data_rank[:21]
```

```python
pd.DataFrame(rank_fix)
```

```python
data_name =[]

for n in range(31, len(data_quote),5):
    data_name.append(data_quote[n].text)
```

```python
data_name
```

```python
pd.DataFrame(data_name)
```

```python
data_world =[]

for w in range(32, len(data_quote),5):
    data_world.append(data_quote[w].text)
```

```python
data_world
```

```python
pd.DataFrame(data_world)
```

```python
data_rise =[]

for ri in range(33, len(data_quote),5):
    data_rise.append(data_quote[ri].text)
```

```python
data_rise
```

```python
pd.DataFrame(data_rise)
```

```python
data_sow =[]

for sw in range(34, len(data_quote),5):
    data_sow.append(data_quote[sw].text)
```

```python
data_sow
```

```python
pd.DataFrame(data_sow)
```

```python
data = {
        'Rank': rank_fix,
        'Name': data_name,
        'Rise/Fall': data_rise,
        'World': data_world,
        'Source of Wealth': data_sow,
    }
```

```python
pd.DataFrame(data)
```

```python
pd.DataFrame(data).to_excel('thetimes.xlsx')
```
